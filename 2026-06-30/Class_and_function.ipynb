## 코틀린의 여러가지 클래스 활용 및 람다 함수 학습

```kotlin
open class Shape{
    open fun draw(){
        println("도형 그리기")
    }
}

open class Circle:Shape(){
    override fun draw(){
        print("Shape을 상속받아: ")
        println("원그리기")
    }
}

class Rect:Shape(){
    override fun draw()=
        println("사각형그리기")
        fun info() {
        println("저는 사각형(Rect) 객체입니다!")
    }
}

class Ellipse: Circle()// 내용이 없으면 {} 생략 가능

val shapes:List<Shape> = listOf(Circle(),Rect(),Ellipse())

for(s in shapes){
    s.draw()
}

shapes.forEach{it.draw()} // 단일 argument인 경우에 it으로 대체 가능(s in shapes -> it)

// shapes[1].info() // 오류 발생, Rect 객체이지만 부모 shape의 참조변수에 할당이 되서 , rect를 모름

println("사각형 검사: ${shapes[1] is Rect}")
if(shapes[1] is Rect){ //현재 객체가 Rect에서 생성이 되었다면
    val r = shapes[1] as Rect // 부모 클래스의 객체 변수를 Rect로 변한(downcasting)
    r.info()
    println(r::class)
}

for(s in shapes){
    when(s){
        is Ellipse ->println("타원객체")
        is Circle ->println("원객체")
        is Rect ->println("사각객체")
        
        is Shape -> println("도형객체")
        else->{
            println("내가 모르는 객체")
        }
    }
}
```

## 추상 클래스 학습

```kotlin
abstract class Animal1 //open키워드가 없어도 상속 가능

abstract class Animal2{ // 추상 메서드가 포함된 ㅋ늘래스는 반드시 추상 클래스로 선언
    abstract fun bark()
    open fun run() = println("달립니다.")//추상클래스의 일반메소드의 경우에는 override하려면 open키워드를 붙여야한다.
}

class Dog:Animal1(){

}
class Cat:Animal2(){//메소드오버로딩
    override fun bark(){
        println("야옹")
    }
    override fun run()=println("야옹이 달린다")
    fun run(name:String) {
        print(name)
        println("이 달립니다.")
        }
}
val dog = Dog()
val cat = Cat()

cat.run()
cat.run("야옹맨")
cat.bark()
```

## 인터페이스학습

```kotlin
interface Flyable{
    fun fly() //추상메서드
    fun land() = println("착륙")// 기본메서드(default)
}

class Duck: Flyable{
    override fun fly() = println("오리가 날다")
    override fun land() = println("오리가 앉다.")
}

val duck = Duck()
duck.land()
duck.fly()
```

```kotlin
interface Printer {
    fun print(msg:String)// 추상 메소드
}

class ColorPrinter: Printer{
    override fun print(msg:String) = println("[color ]$msg")
}
```

```kotlin
class Logger1: Printer{
    override fun print(msg:String) = println("[BW] $msg")
    fun log(msg: String){
        print(msg)
    }
}

val logger1 = Logger1()
logger1.log("시작")
```

```kotlin
val color_printer = ColorPrinter()
color_printer.print("시작")
```

```kotlin
class Logger2(private val printer: Printer): Printer by printer{//클래스 위임, 추상메소드 구현을 이미 구현한 다른 클래스에게 위임
    fun log(msg: String){
        print(msg)
    }
}

val logger2 = Logger2(color_printer)
logger2.log("시작")
```

## 람다 함수

```kotlin
val double : (Int)->Int ={it *2}
val d = double(4)
println(d)
```

```kotlin
fun multiplier(factor:Int):(Int)->Int={it*factor}// 입력값에 factor를 곱해서 반환하는 함수
val unknown =multiplier(3)
println("함수 동작 : ${unknown(5)}")
```

```kotlin
val some = {no:Int -> println("$no 값 입력")}
some(10)
```

```kotlin
val some2 : (Int)->Unit = {println("$it 입력")}
some2(10)
```

```kotlin
fun testFun(no:Int, arg: (Int)->Boolean): ()->String{
    val result = if(arg(no)){//(Int)->Boolean,trun/false
        "valid"
    }
    else{
        "invalid"
    }
    return { "testFun result: $result"}//{()->"..."}=>{->"..."}=>{"..."}
}

val testResult = testFun(10, {no -> no>0})
testResult()
```
