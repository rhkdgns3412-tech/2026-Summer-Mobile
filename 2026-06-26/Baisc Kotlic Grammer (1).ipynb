## 함수와 데이터타입

- Unit 반환 문 생략 가능

```kotlin
fun add(a:Int, b:Int): Int{
    return a+b
}

println("2+3= ${add(2,3)}")

fun showMessage(msg: String):Unit{
    println(msg)
}

showMessage("안녕하세요")
```

## 함수의 기본값 정의
- 함수 인수에 값을 지정하면, 가변 인수를 전달할수 있음

```kotlin
import kotlin.math.pow
fun power(base:Double, exp:Int =2):Double{
    return base.pow(exp)
}

println("2^3 = ${power(2.0,3)}")
println("3^2 = ${power(3.0)}")
```

## when 제어문 사용
- when 제어문은 기존의 swtich 문보다 강력함
- 일반적인 swtich은 if 동등절에 해당하나, when 제어문은 범위 및 다양한 조건을 포함할 수 있음

```kotlin

val data : Any = "Test"
when(data){
    is String -> println("입력이 String임.")
    20,30 -> println("입력값이 20 또는 30임")
    in 1..10 -> println("입력 값이 1이상 10 이하")
    else -> println("입력값이 예상 범위를 벗어남")

}
```

## 반복문을 통해서 원주율 계산하기
- pi/4 = 1-1/3+1/5-1/7

```kotlin
fun series(n:Int):Double{
    var result:Double =0.0
    for(i in 1..n){
        if(i%2==0)
            result-=(1.0/(i*2-1))
        else
            result+=(1.0/(i*2-1))
    }
    result*=4.0
    return result
}
println("${series(100)}")
println("${series(500)}")
println("${series(1000)}")
```
